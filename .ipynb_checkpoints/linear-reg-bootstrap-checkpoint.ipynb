In the notebook, we train a linear model with a decision-maker for highly non-linear data to show that it can learn to directly map the non-linear data, without conforming to the linear model. We demonstrate that a sufficiently complex decision-maker may learn to surpass model predictions, but the effect can be alleviated with regularization.

### Auxiliary functions for handling Stan objects

```python
def pystan_vb_extract(results):
    param_specs = results['sampler_param_names']
    samples = results['sampler_params']
    n = len(samples[0])

    # first pass, calculate the shape
    param_shapes = OrderedDict()
    for param_spec in param_specs:
        splt = param_spec.split('[')
        name = splt[0]
        if len(splt) > 1:
            idxs = [int(i) for i in splt[1][:-1].split(',')]  # no +1 for shape calculation because pystan already returns 1-based indexes for vb!
        else:
            idxs = ()
        param_shapes[name] = np.maximum(idxs, param_shapes.get(name, idxs))

    # create arrays
    params = OrderedDict([(name, np.nan * np.empty((n, ) + tuple(shape))) for name, shape in param_shapes.items()])

    # second pass, set arrays
    for param_spec, param_samples in zip(param_specs, samples):
        splt = param_spec.split('[')
        name = splt[0]
        if len(splt) > 1:
            idxs = [int(i) - 1 for i in splt[1][:-1].split(',')]  # -1 because pystan returns 1-based indexes for vb!
        else:
            idxs = ()
        params[name][(..., ) + tuple(idxs)] = param_samples

    return params
```

```python
# to avoid recompiling unchanged models
def StanModel_cache(model_code, model_name=None, **kwargs):
    """Use just as you would `stan`"""
    code_hash = md5(model_code.encode('ascii')).hexdigest()
    if model_name is None:
        cache_fn = 'cached-model-{}.pkl'.format(code_hash)
    else:
        cache_fn = 'cached-{}-{}.pkl'.format(model_name, code_hash)
    try:
        sm = pickle.load(open(cache_fn, 'rb'))
    except:
        sm = pystan.StanModel(model_code=model_code)
        with open(cache_fn, 'wb') as f:
            pickle.dump(sm, f)
    else:
        print("Using cached StanModel")
    return sm
```

### Stan Bayesian linear regression model

```python
model = """
data {
  int<lower=0> N;
  vector[N] x;
  vector[N] y;
}
parameters {
  real alpha;
  real beta;
  real<lower=0> sigma;
}
model {
  y ~ normal(alpha + beta * x, sigma);
}
"""
```

```python
stanmodel = StanModel_cache(model_code = model, model_name="model")
```

```python
np.random.seed(123)
def make_data(N, D):
    x_data = np.float32(np.random.uniform(-10.5, 10.5, (1, N))).T
    r_data = np.float32(1. * np.random.normal(size=(N,1)))
    y_data = np.float32(np.sin(0.75*x_data)*7.0+x_data*0.5+r_data*3.0)
    return x_data, y_data
x_train, y_train = make_data(500, 1)
```

```python
x_train = (x_train - x_train.mean(axis = 0))/x_train.std(axis = 0)
y_train = (y_train - y_train.mean(axis = 0))/y_train.std(axis = 0)
plt.plot(x_train[:,0], y_train, '.');
```

```python
iterations = 2000
batch_size = 32
q_y = 0.9
num_examples = x_train.shape[0]

def tilted_loss(q,y,f):
    e = (y-f)
    return K.mean(K.maximum(q*e, (q-1)*e), axis=-1)

def tilted_optimal_h(ys, q):   
    quantiles = np.percentile(ys, int(q*100), axis=0)
    return quantiles

def tilted_loss_numpy(q,y,f):
    e = (y-f)
    return np.mean(np.maximum(q*e, (q-1)*e), axis=-1)
```

```python
def create_data():

  data = {'N': x_train.shape[0],
          'x': x_train.reshape(-1),
          'y': y_train.reshape(-1)
         }
  return data
```

```python
def create_data_bootstrap():
    idx = np.random.choice(np.arange(len(y_train)), len(y_train), replace = True)
    data = {'N': x_train[idx,:].shape[0],
            'x': x_train[idx,:].reshape(-1),
            'y': y_train[idx,:].reshape(-1)
            }        
    return data
```

```python
alpha = []
beta = []
sigma = []
nbootstrap = 10
for b in range(nbootstrap):
    stanfit = stanmodel.sampling(data = create_data_bootstrap(), chains = 5)
    alpha.append(stanfit.extract('alpha')['alpha'])
    beta.append(stanfit.extract('beta')['beta'])
    sigma.append(stanfit.extract('sigma')['sigma'])
```

```python
nsamples = 5000
ys = np.zeros([nbootstrap, nsamples, x_train.shape[0]])
for b in range(nbootstrap):
    for i in range(nsamples):
        aa, bb, ss = (alpha[b])[i], (beta[b])[i], (sigma[b])[i]
        ys[b,i]= np.random.normal(bb * x_train + aa, ss).reshape(-1)
```

```python
def tilted_loss_numpy(q,y,f):
    e = (y-f)
    return np.maximum(q*e, (q-1)*e)
```

### Bootstrap priors for the decision $h$

```python
h_b = []
for b in range(nbootstrap):
    h_b.append(tilted_optimal_h(ys[b], 0.9))
h_b = np.array(h_b)
h_scale = 1.0/(2. * h_b.std(0)**2)
h_loc =  h_b.mean(0)
```

```python
plt.plot(x_train[:,0], y_train[:,0], '.')
p = np.argsort(x_train[:,0])
for b in range(nbootstrap):
    plt.plot(np.sort(x_train[:,0]), h_b[b][p])
```

```python
stanfit = stanmodel.sampling(data = create_data(), chains = 5)
alpha = (stanfit.extract('alpha')['alpha'])
beta = (stanfit.extract('beta')['beta'])
sigma = (stanfit.extract('sigma')['sigma'])
```

```python
nsamples = 5000
ys = np.zeros([nsamples, x_train.shape[0]])
for i in range(nsamples):
    aa, bb, ss = (alpha)[i], (beta)[i], (sigma)[i]
    ys[i]= np.random.normal(bb * x_train + aa, ss).reshape(-1)

y_m, y_err = np.mean(ys, 0), np.std(ys, 0)
h =  tilted_optimal_h(ys, q_y)
```

```python
ys_train = np.zeros([nsamples, x_train.shape[0]])
for i in range(nsamples):
    aa, bb, ss = (alpha)[i], (beta)[i], (sigma)[i]
    ys_train[i]= np.random.normal(bb * x_train + aa, ss).reshape(-1)
y_train_quantiles = np.array([np.percentile(ys_train, int(q*100), axis=0) for q in np.arange(0,1.01,0.05)]).T
augmented_y_train = np.hstack([y_train, h_loc.reshape(-1,1), h_scale.reshape(-1,1)])
```

## DM training with $q$ predictive

```python
lams = [0.0,  0.0005, 0.01]
h_lam = []

for l in lams:
    def risk(y_true, y_pred):
        y_t = K.reshape(y_true[:,0],(-1, 1))
        h_m = K.reshape(y_true[:,1],(-1, 1))
        h_s = K.reshape(y_true[:,2],(-1, 1))
        return tilted_loss(0.9, K.reshape(y_true[:,0],(-1, 1)), y_pred)   +  l * h_s * (y_pred - h_m)**2

    x_input =  tf.keras.Input((y_train_quantiles.shape[1:]), name = 'x_input')
    x = tf.keras.layers.Dense(20, activation = 'relu')(x_input)
    x = tf.keras.layers.Dense(10, activation = 'relu')(x)
    y_output = tf.keras.layers.Dense(1)(x)

    model = tf.keras.Model(x_input, y_output)
    model.compile(optimizer = tf.train.AdamOptimizer(0.01), loss = risk ) 
    model.fit(x = y_train_quantiles, y = augmented_y_train, batch_size=32, epochs = 500, verbose = 0)
    y_s = model.predict(y_train_quantiles)
    h_lam.append(y_s)
```

```python
lams_str = ['no reg.', r'boot. $(\lambda = 5\times10^{-4})$', r'boot. $(\lambda = 1\times10^{-2})$']
def errorfill(x, y, yerr, y_train, h, h_lams, p, color=None, alpha_fill=0.2, ax=None, label="", lw=2, ls="-"):
    ax = ax if ax is not None else plt.gca()
    if color is None:
        color = ax._get_lines.color_cycle.next()
    if np.isscalar(yerr) or len(yerr) == len(y):
        ymin = y - yerr
        ymax = y + yerr
    elif len(yerr) == 2:
        ymin, ymax = yerr
    tck, u = splprep([x, y], s=10)
    new_points = splev(u, tck)        
    ax.plot(new_points[0], new_points[1], color=RED, alpha = 1., label= "Mean prediction", lw=1, ls='--')
    tck, u = splprep([x, h], s=10)
    new_points = splev(u, tck)
    ax.plot(new_points[0], new_points[1], color=RED, alpha = 1., label="q-optimal", lw=3, ls=ls)    
    
    ax.fill_between(x, running_mean(ymax), running_mean(ymin), color=RED, alpha=alpha_fill, linewidth=0.0)
    alphas = ['--','-.',':']
    cols = ['dodgerblue', 'steelblue', 'gray'][::-1]
    for a,b,c,d in zip(lams_str, h_lam, alphas, cols):
        tck, u = splprep([x, b[p].reshape(-1)], s=10)
        new_points = splev(u, tck)
        ax.plot(new_points[0], new_points[1], alpha = 1., color = 'b', label = r"DM " + (a), lw=2.5, ls=c)      
    ax.plot(x[::3], y_train[::3], '.',  color=RED, alpha = 0.2)
```
